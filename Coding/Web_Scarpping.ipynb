```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import urllib.parse 
import urllib.request
```

```python
names = []
prices = []           
descriptions = []
websites = []
```

```python
root_url = 'https://jmty.jp/'
relative_url = 'saitama/sale-bic/g-all/a-179-saitama/p-'
url_combined = []

for link in relative_url:
  url_combined.append(urllib.parse.urljoin(root_url,link))
```

```python
for i in range (1,5):
  website = 'https://jmty.jp/saitama/sale-bic/g-all/a-179-saitama/p-' + str(i)
  response = requests.get(website)
  soup = BeautifulSoup(response.content,'html.parser')
  items = soup.find_all('li',{'class': 'p-articles-list-item'})
  for  item in items:

    try:
        names.append(item.find('div',{'class':'p-item-title'}).get_text().strip())
    except:
        names.append("n/a")

    try:
        prices.append(item.find('div',{'class':'p-item-most-important'}).get_text().strip())
    except:  
        prices.append("n/a")

    try:
        descriptions.append(item.find('div',{'class':'p-item-detail'}).get_text().strip())
    except:
        descriptions.append("n/a")

    try:
        way_to_link = item.find('div',{'class':'p-item-title'})
        websites.append(way_to_link.find('a').get('href'))
    except:
        websites.append('n/a')
```

```python
product_overview = pd.DataFrame({"Name":names, "Prices":prices, "Descriptions": descriptions, "Website": websites})
product_overview
```

```python
product_overview.to_excel('Web_Scrapping_to_find_Bicycle_Saitama_Jimoti.xlsx', index=False)
```
